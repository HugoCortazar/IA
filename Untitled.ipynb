```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="white")
```

```python
df = pd.read_csv("simulation_results_20250404_183257.csv")
```

```python
print("Dimensiones del dataset:", df.shape)
print("\nPrimeras filas:")
display(df.head())

# Estadisticas descriptivas
print("\nEstadisticas descriptivas:")
display(df.describe(include='all').round(2))
```

```python
# Tiempos de procesamiento
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Component', y='Total_Time')
plt.title('Distribucin de tiempos totales por componente')
plt.xticks(rotation=45)
plt.ylabel('Tiempo Total')
plt.show()
```

```python
# Retrasos en transporte
plt.figure(figsize=(10,5))
sns.histplot(df['Transport_Delay'], bins=20, kde=True)
plt.title('Distribucin de retrasos en trasnsporte')
plt.xlabel('Retraso (unidades de tiempo)')
plt.show()
```
